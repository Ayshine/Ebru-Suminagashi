```python
# Import libraries necessary for this project
import pandas as pd
import matplotlib.pyplot as plt
# Pretty display for notebooks
%matplotlib inline
```

```python
df = pd.read_csv('results.csv')
```

```python
from IPython.display import display
display(df.head())
```

```python
tl_models = ["AlexNet","DenseNet", "VGG","VGG_BN"]

fs_models = ["Simple_CNN_Resize", "Simple_CNN_CenterCrop", 
             "Multilayer_CNN_128_Resize","Multilayer_CNN_256_Resize", 
             "Multilayer_CNN_128_CenterCrop","Multilayer_CNN_256_CenterCrop"]
```

```python
fig, ax = plt.subplots()

for model in tl_models:
    sub_df = df[df.Model_Name==model][['Epoch', 'Test_Accuracy']]
    sub_df.rename({'Test_Accuracy': model}, axis=1, inplace=True)
    sub_df.plot(x='Epoch', y=model, ax=ax)

# Shrink current axis by 20%
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
fig, ax = plt.subplots()

for model in fs_models:
    sub_df = df[df.Model_Name==model][['Epoch', 'Test_Accuracy']]
    sub_df.rename({'Test_Accuracy': model}, axis=1, inplace=True)
    sub_df.plot(x='Epoch', y=model, ax=ax)

# Shrink current axis by 20%
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
